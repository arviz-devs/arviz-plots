# Examples with `plot_posterior`

```python
from arviz_plots import plot_posterior
```

```python
from arviz_base import load_arviz_data

post = load_arviz_data("centered_eight").posterior.ds
```

```python
plot_posterior(post);
```

```python
plot_posterior(post, plot_kwargs={"kde": {"color": "orange"}});
```

```python
plot_posterior(post, pc_kwargs={"aes": {"color": ["school"]}, "color": [f"C{i}" for i in range(8)]});
#TODO: moodificar el aes mapping per tal que noms s'utilitzi si la dimensi existeix en aquella variable
#TODO: crear atribut public per aes_keys i per totes les dimensions que tenen algun aesthetics assignat
```

```python
plot_posterior(
    post, 
    pc_kwargs={"aes": {"color": ["school"], "ls": ["chain"]}, "color": [f"C{i}" for i in range(8)], "ls": ["--", ":"], "plot_grid_kws": {"figsize": (12, 7)}},
    aes_map={"kde": ["color", "ls"], "point_estimate": ["color"]},
    plot_kwargs={"ci": {"color": "lime"}}
);
```

```python
plot_posterior(
    post, 
    pc_kwargs={"cols": ["__variable__"], "aes": {"color": ["school"]}, "color": [f"C{i}" for i in range(8)], "ls": ["--", ":"], "plot_grid_kws": {"figsize": (12, 7)}},
    aes_map={"kde": ["color", "ls"], "point_estimate": ["color"]},
    plot_kwargs={"ci": {"color": "lime"}},
);
```
